```python
# Lets replace the range data into mean
```

```python
avg_wbn=(300+350)/2
avg_karn=(5648+6488)/2
avg_kerl=(5942+6422)/2

avg_t1=27669+27719/2
avg_t2=29391+30711/2

data=data.replace('300-350',avg_wbn)
data=data.replace('5648-6488',avg_karn)
data=data.replace('5942-6422',avg_kerl)

data=data.replace('27669-27719',avg_t1)
data=data.replace('29391-30711',avg_t2)

```

```python
data
```

```python
data.info()
```

```python
for feature in data:
    data[feature]=data[feature].astype("float")
```

```python
# replace the nan columns with mean
for feature in data:
    median_value=data[feature].median()
    data[feature].fillna(median_value,inplace=True)
```
